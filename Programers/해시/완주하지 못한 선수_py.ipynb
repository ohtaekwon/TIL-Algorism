## 문제
수많은 마라톤 선수들이 마라톤에 참여하였습니다. 단 한 명의 선수를 제외하고는 모든 선수가 마라톤을 완주하였습니다.

마라톤에 참여한 선수들의 이름이 담긴 배열 participant와 완주한 선수들의 이름이 담긴 배열 completion이 주어질 때, 완주하지 못한 선수의 이름을 return 하도록 solution 함수를 작성해주세요.

제한사항
마라톤 경기에 참여한 선수의 수는 1명 이상 100,000명 이하입니다.
completion의 길이는 participant의 길이보다 1 작습니다.
참가자의 이름은 1개 이상 20개 이하의 알파벳 소문자로 이루어져 있습니다.
참가자 중에는 동명이인이 있을 수 있습니다.


## 제한사항

- 마라톤 경기에 참여한 선수의 수는 1명 이상 100,000명 이하입니다.
- completion의 길이는 participant의 길이보다 1 작습니다.
- 참가자의 이름은 1개 이상 20개 이하의 알파벳 소문자로 이루어져 있습니다.
- 참가자 중에는 동명이인이 있을 수 있습니다.

## 입출력 예

|participant|completion|return|
|:--|:--|:--:|
|["leo", "kiki", "eden"]|["eden", "kiki"]|"leo"|
|["marina", "josipa", "nikola", "vinko", "filipa"]|["josipa", "filipa", "marina", "nikola"]|"vinko"|

## 풀이 방법 1
 
- 해시를 이용해보기

#### 문제 접근 방법

- 제한사항에서 요구조건을 파악
- 참가자는 100,000명 이하
- 참가자 중에 동명이인이 있을 수 있다.
- 이름이 주어지면 몇 번이나 배열에 등장했는지, 수를 기록하고 저장할 수 있는 것을 원함.
    - 해쉬와 연관

#### 자료구조(와 알고리즘)의 선택

- 만약 이름 대신 번호가 주어졌다면?
    - 선형 배열(linear array)
- 번호 말고 다른 것(예: 문자열)로 접근할 수 있는 좋은 자료구조는 없나?

#### 해시(Hash)
- 사람들의 이름을 `키(Key)`로 지정하고, 저장공간 내에다 어느 위치에 있을지 정해서해시 테이블(Hash Table)에 `값(value)`으로 저장
- 각각 해당하는 값이 구성하기 위해서 `hash function`이 적용된다.
    - 가급적이면 서로 다른 칸에 넣을 수 있도록
    - 같은 칸에 넣으면 충돌(collision)이 발생
- 해쉬테이블 내 각각의 칸
    - 해쉬 버킷(hash bucket)

#### 문제 풀이의 흐름

- `participant` 라는 배열에 나타난 이름들을 살피면서, 이름이 나타난 횟수를 샌다.
    - 이름을 만났을 때, 해당 이름에 `1` 을 대응시키는 해쉬테이블을 만든다.
    ```
     ex) mislav → 1
          stanko → 1
          mislav → 2
          ana → 1
    ```
    - 대응된 관계를 해쉬테이블에 기록
    
- 그다음, `completion` 라는 배열에 나타난 이름들을 살핀다.
    - 이름을 만났을 때, 해당 이름에 `-1` 을 대응시키는 해쉬테이블을 만든다.
    ```
     ex) mislav → 1
          stanko → 0
          ana → 0
    ```
- 지금 해쉬테이블에 등록된 이름 중에 0이 아닌 값이 있게 된다.
    - 즉, 1을 가진 이름이 완주하지 못한 사람이 된다.
    
    
### Python 사전 (dictionary)

```python
d={"leo":30, "kiki":62}

x=d["leo"]
d["leo"] = 58
```
- 사전의 원소들을 해시를 이용해 $O$(1) 시간에 접근이 가능하다는 특징이 있다.


````python
def solution(participant, completion):
    # 빈 사전을 만들고
    d = {}
    
    for x in participant:
        # 등장하는 x에 대해서 get메서드를 활용해,
        # x가 존재하면 x , 없으면, 0을 반환
        # 처음 등장이면 1
        d[x]=d.get(x,0) + 1
    for x in completion:
        d[x]-=1
    
    dnf= [k for k, v in d.items() if v>0]
       
    return dnf[0]
````

- 사전은 해쉬테이블로 구성되어있기 때문에, 상수시간안에 발생한다.
- 따라서, 
    ```python
    for x in participant:
        [x]=d.get(x,0) + 1
    ```
    이 순환문의 시간 복잡도는 participant라는 배열의 길이에 비례한다.
- 또한, 
    ```python
    for x in completion:
        d[x]-=1
    ```
    completion의 배열의 길이에 비례한다.
    
- 하지만, `dnf= [k for k, v in d.items() if v>0]` 은 사전에 들어 있는 모든 원소를 꺼내서 비교하는 것이기 때문에, 사전의 크기에 비례하는 복잡도를 가진 연산이다.

- 이 함수의 전체 시간 복잡도는 `participant` 배열의 길이를 n이라 하고, 
- n에 비례하는 linear time 알고리즘이다.

## 풀이 방법 2

- 정렬을 이용한다면?
- 각 리스트들을 정렬한다.
- `completion`을 반복하여 실행하며 `participant`와 비교한다.
- 같지 않다면, 같지 않은 요소를 리턴한다.
- 같다면, 멘 마지막 요소를 출력한다.

````python
def solution(participant, completion):
    # 각 리스트들을 정렬시켜준다. 
    participant.sort()
    completion.sort()


    for i in range(len(completion)): # completion 안의 리스트 길이만큼 반복
        if participant[i]!=completion[i]: # 정렬된 리스트들이 같지 않다면
            return participant[i] # 같지 않은 요소를 리턴
    # 해당 for문을 다 돌았을 경우 리턴이 되지 않았다면, 완벽하게 같았다는 의미
    # 또한, 완주하지 못한 사람은 무조건 1명이라고 했으니, participant 보다 completion 가 언제나 1명씩 더 많다. 
    # 아직 비교하지 못한 요소는 participant 리스트의 맨 뒤의 요소이므로,
    return participant[-1] # participant의 맨 마지막 남은 요소(participant 인덱스 끝)이 리턴되어야 한다.
````

- sort()의 최적 알고리즘은 $O$($N$log$N$)에 비례한다.
- 정렬을 이용하면, 시간 복잡도가 테스트는 통과하지만, 문제의 요점은 hash를 이용하는 것에 맞다.


### 풀이방법 3

- `collections` 모듈에서 `Counter` 클래스를 사용

````python
from collections import Counter

def solution(participant, completion):
    # counter 클래스를 적용
    counter_participant = Counter(participant)
    counter_completion = Counter(completion)
    
    # counter 끼리 연산을 사용
    last = counter_participant - counter_completion
    
    # 값이 남은 하나가 완주하지 못한 인원이 된다. 
    last = list(last.keys())[0]
    return last
````
