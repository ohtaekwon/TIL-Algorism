## 문제

선후수 수강(레벨3)

윤서는 자신이 수강하고자 하는 과목을 신청하기 위해서 수강편람을 읽었고, 어느 과목을 수강하기 위해서는 그 이전에 이미 수강했어야 하는 과목들이 정해져있음을 알았습니다. 예를 들어 과목 "A"를 수강하기 위해서는 그 앞서 과목 "B"를 수강해야 한다면 과목 "B"를 과목 "A"의 선수과목이라고 부릅니다.

윤서가 특정한 과목 "Z"를 수강하기 위해서는 이 과목의 선수과목인 "Y"를 먼저 이수해야 하며, 과목 "Y"에도 선수과목 "X"가 지정되어 있다면 과목 "X"를 이수한 후에야 과목 "Y"를 수강신청하여 이수할 수 있습니다.

윤서가 수강하고자 하는 과목을 정했을 때, 이 과목을 수강하기 위해서는 적어도 어떤 과목들을 어떤 순서에 따라 수강해야 하는지를 알아내는 함수 `solution()`을 완성하세요.

함수 `solution()`의 입력으로는 다음과 같은 것들이 주어집니다.

- 선후수 관계를 나타내는 배열 `s1`과 `s2` : 배열 `s1`의 `i`번째 원소가 `"X"` 이고 배열 `s2`의 `i`번째 원소가 `"Y"` 인 경우, 과목 "X"가 과목 "Y"의 선수과목임을 나타냅니다.
- 윤서가 수강하고자 하는 과목 코드 `k`

함수 `solution()`은 `k`에 의하여 주어진 코드를 가지는 과목을 이수하기 위하여 순서대로 이수해야 하는 과목들의 코드를 나열한 배열 `answer`를 반환합니다.

## 제한사항

- 수강편람에 기재된 과목의 수는 8이상 10,000 이하입니다.
- 선후수 관계를 나타내는 배열 `s1`과 `s2`의 길이는 서로 같으며, 9이상 100,000 이하입니다.
- 각 과목에는 수강 코드가 부여되어 있으며, 수강 코드는 10글자 이내의 영문 알파벳 대문자로 이루어져 있습니다. 
- 어떠한 과목 "X"와 "Y"사이에도 순환적 선후수 관계는 주어지지 않습니다. 즉, 과목 "X"를 수강하기 위하여 과목 "Y"를 먼저 이수해야 하는데 과목 "Y"의 선수과목 관계를 따라 보았을 때, 이 과목을 수강하기 위해서는 "X"를 우선 이수했어야 하는 경우는 없습니다.
- 서로 선후수 관계가 없는 과목 "M"과 "K"를 이수하는 순서는 과목 코드의 알파벳 순서에 따른다고 가정합니다.

### deque의 의미

- `큐(queue)` 선입선출(FIFO) 방식으로 작동하는 반면에 덱(deque)은 `양방향 큐`처럼 작동한다.
- 즉, 양쪽 방향에서 원소를 추가하거나 삭제할 수 있다.
- `deque`는 양 끝 원소의 append, pop이 압도적으로 빠르다($0$(1) 상수시간 소요)
    - 반면, list는 양 끝 원소에 접근하여, 삽입, 삭제 하려고 하면 $O(N)$ 시간이 소요된다.
    
### deque는 언제 사용하는가?

- 덱은 스택(stack)처럼 사용할 수도, 큐(queue)처럼 사용할수도 있다.
- 시작점의 값을 넣고, 빼거나 끝 점의 값을 넣고 뺴는 데에 최적화된 연산 속도를 제공한다.
- 대부분의 겨웅 덱이 리스트보다 월등한 옵션을 가지고 있다.
- 덱은 특히 `push/pop` 연산이 빈번한 알고리즘에서 장점이 부각된다.

````python
# 선수과목에 맞게 배열을 나타내기
# S1의 길이와 S2의 길이는 같다
# BFS로 풀이
from collections import defaultdict

def solution(s1, s2, k):

    answer = []

    # 선수강 강의가 valeu로, 후수강 강의가 node로 하는 dict
    lecture=defaultdict(list)
    for idx2, j in enumerate(s2):
        for idx1,i in enumerate(s1):
            if idx2==idx1: 
                lecture[j].append(i)
    visited = list()
    need_visit = list()
    none_pre=list() # 선수강 강의가 없는 경우
    need_visit.append(k) # 수강해야할 강의를 먼저 넣고
    while need_visit: 
        node = need_visit.pop(0) # B를 node로 하고
        if node not in visited : 
            if lecture[node]:
                visited.append(node)
                need_visit.extend(lecture[node])
            elif lecture[node]!=None: # 선수강 강의가 없는 경우
                none_pre.append(node)

    none_pre=list(set(none_pre)) # 중복 제거
    none_pre=sorted(none_pre, reverse=True) # 알파벳순 정렬
    visited.extend(none_pre)
    return visited[::-1]
````

````python
from collections import deque
import heapq

class lacture:
    def __init__(self, data):
        self.preLactures = set()
        self.name = data
        self.check = False

    def add_preLactures(self, lactures):
        self.preLactures.add(lactures)

    def erase_preLactures(self, lactures):
        self.preLactures.discard(lactures)

def getLactureTable(s1, s2, k):
    lactures = dict()

    for index, prolacture in enumerate(s2):
        if not prolacture in lactures:
            lactures[prolacture] = lacture(prolacture)
        if not s1[index] in lactures:
            lactures[s1[index]] = lacture(s1[index])

        lactures[prolacture].add_preLactures(s1[index])

    ret = dict()
    q = deque()
    ret[k] = lactures[k]
    q.append(lactures[k])

    while q:
        temp = q.popleft()
        for pre in temp.preLactures:
            if pre in ret:
                continue

            ret[pre] = lactures[pre]
            q.append(lactures[pre])


    return ret


def topological(lacturesList, k):
    ret = []
    pq = []

    for lacture in lacturesList.values():
        if not lacture.preLactures:
            heapq.heappush(pq, lacture.name)
            lacture.check = True

    while pq:
        temp = []
        ret.append(pq[0])

        if ret[-1] == k:
            break

        for lacture in lacturesList.values():
            if ret[-1] in lacture.preLactures:
                lacture.preLactures.remove(ret[-1])
            if not lacture.check and not lacture.preLactures:
                temp.append(lacture.name)
                lacture.check = True

        heapq.heappop(pq)
        for item in temp:
            heapq.heappush(pq, item)

    return ret


def solution(s1, s2, k):
    answer = []

    # 문제의 조건에서 scc가 없음이 확정되었으므로
    # 단순한 위상정렬로 해결가능
    lacturesList = getLactureTable(s1, s2, k)
    answer = topological(lacturesList, k)

    return answer
````

````python
from collections import deque
from collections import defaultdict
from collections import OrderedDict
import heapq

def dfs_search(current, graph, visited, k):
    visited[current] = True
    found = False
    for x in graph[current]:
        if x == k:
            return True
        if not visited[x]:
            found = dfs_search(x, graph, visited, k)
            if found:
                break
    return found

def topology_dfs(current, graph, indegree, result, k):
    result.append(current)
    if current == k:
        return
    while graph[current]:
        nxt = heapq.heappop(graph[current])
        indegree[nxt] -= 1
        if indegree[nxt] == 0:
            topology_dfs(nxt, graph, indegree, result, k)
    
def solution(s1, s2, k):
    graph = defaultdict(list)
    indegree = OrderedDict()
    result = []
    
    for i, v in enumerate(s1):
        w = s2[i]
        heapq.heappush(graph[v], w)

        if v not in indegree:
            indegree[v] = 0
        
        if w not in indegree:
            indegree[w] = 0

        indegree[w] += 1
    
    for i, v in indegree.items():
        if v == 0:
            visited = defaultdict(bool)
            found = dfs_search(i, graph, visited, k)
            if (found):
                topology_dfs(i, graph, indegree, result, k)

    return result
````
