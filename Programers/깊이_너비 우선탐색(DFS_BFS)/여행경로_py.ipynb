## 문제
아래와 같이 5와 사칙연산만으로 12를 표현할 수 있습니다.

12 = 5 + 5 + (5 / 5) + (5 / 5)
12 = 55 / 5 + 5 / 5
12 = (55 + 5) / 5

5를 사용한 횟수는 각각 6,5,4 입니다. 그리고 이중 가장 작은 경우는 4입니다.
이처럼 숫자 N과 number가 주어질 때, N과 사칙연산만 사용해서 표현 할 수 있는 방법 중 N 사용횟수의 최솟값을 return 하도록 solution 함수를 작성하세요.

### 깊이 우선 탐색(DFS; Depth-First-Search)

- 한 정점에서 인접한 모든(아직 방문하지 않은)정점을 방문하되, 
- 각 인접 정점을 기준으로 깊이 우선 탐색을 끝낸 후 다음 정점으로 진행

#### 예시

### 너비 우선 탐색(BFS; Breadth-Frist-Search)

- 한 정점에서 인접한 모든(아직 방문하지 않은)정점을 방문하고, 
- 방문한 각 인접 정점을 기준으로 (방문한 순서에 따라) 또다시 너비 우선 탐색을 행함
- 큐를 이용하여 어느 정점에서 BFS를 해야하는지를 기록하고 행함

### 알고리즘의 설계

- ICN → ATL 선택 후 없는 것으로 취급
- ATL → ICN(알파벳 순서 우선) 으로 선택 후 빼기
- 이와 같은 패턴 반복한다.
- 문제가 재귀적인 성질이 있다.
- 스택을 이용하여 재귀적인 "한 붓 그리기"문제를 해결
    → DFS알고리즘 응용
    
### 알고리즘의 동작

- ICN → AAA 
- AAA의 다음 행선지가 없으면 꺼내서 다른 곳에 적어둔다
- 다시 ICN → BBB
- BBB → CCC
- CCC → BBB
- BBB → ICN (더이상 갈 행선지가 없다.)
- 이제 스택에서 꺼낸다.
- ICN → BBB → CCC → BBB → ICN → AAA (방문한 공항들의 역순)

### 처음 풀이

```python
# a -> b
# 데이터는 10000개까지
# 항상 ICN에서 출발
from collections import defaultdict

def solution(tickets):
    graph = defaultdict(list)
    visited=list()
    need_visit=list()
    # 딕셔너리로 만들기
    name=list(set(sum(tickets,[])))
    for n in name:
        for i in range(len(tickets)):
            if n==tickets[i][0]:
                graph[n].append(tickets[i][1])
    # 알파벳순 정렬
    for g in graph:
        graph[g].sort(reverse=True)


    start='ICN'# 처음에 need_visit에 'ICN'를 넣는다.
    need_visit.append(start)

    while need_visit:
        node=need_visit.pop(0) # 맨 앞의 인자를 빼고
        if node not in visited:
            visited.append(node)
            need_visit.extend(graph[node])
        else:
            visited.append(graph[node][-1])
            # graph[node]=graph[node][-1]
    return visited
```

### 코드 풀이

#### 그래프의 표현
- 사전을 이요하여 각 공항에서 출발하는 항공권의 집합을 표현
- ex) 
    - ICN → {ATL, SFO}
    - ATL → {ICN, SFO}
    - SFO → {ATL}
    - 알파벳 순서가 중요하므로, **리스트**로 표현하는 것이 좋다.
    - 근데, 리스트에서 앞에서부터 원소를 제거하는 것보다 뒤에서 부터 제거하는 것이 더 효율적이다.
    - 역순으로 정렬한다.
        - ICN → [SFO,ATL] / ATL → [SFO,ICN] / SOF → [ATL]
        - 알파벳 역순으로 정렬!

```python
def solution(tickets):
    routes={} # 표들의 사전을 정의
    
    for t in tickets: # 티켓을 꺼내서, 원소는 리스트인데, 앞이 출발 공항, 뒤가 도착 공항
        # 리스트 병합
        # 출발 공항 = routes에서 처음 등장하면 [] 빈리스트를 호출하고,
        # [t[1]] 도착 공항을 리스트 병합한다.
        routes[t[0]] = routes.get(t[0],[]) + [t[1]]
    
    # 알파벳 순서로 정렬
    for r in routes:
        routes[r].sort(reverse=True)
    
    # 스택을 처음 값을 ICN을 넣는다.
    stack = ['ICN']
    
    # 갈려는 경로
    path=[] # 빈리스로 초기화
    
    while len(stack)>0: # stack의 값이 다 없어 질 때까지,
        top = stack[-1] # stack의 맨 위의 원소
        
        # top이 rouutes에 없거나, 또는 있었는데 없게 됐다면,
        if top not in routes or len(routes[top])==0:
            # path에 stack의 맨 앞의 값을 넣는다.
            path.append(stack.pop())
        else: # 그렇지 않으면
            stack.append(routes[top][-1]) # 갈수 있는 공항들 중 마지막 번째 공항
            routes[top] = routes[top][:-1] # 
    return path[::-1] # 역순으로 다시 정렬
```
